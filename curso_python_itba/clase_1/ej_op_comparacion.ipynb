```python
x = int(input("Ingrese el primer número:"))
y = int(input("Ingrese el segundo número:"))
z = ( x> 15 or x <-15) and ( y > 15 or y < -15 )
print("El valor absoluto de ambos numeros es mayor a 15?", z)
```
