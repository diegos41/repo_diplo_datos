```python
#Pruebo código del teórico
def chequarContraseña(c):
    if c == "Secreto":
        resultado = True
    else:
        resultado = False
    return resultado


ingresado = input("Ingrese contraseña: ")        
while not chequarContraseña(ingresado):
  ingresado = input("Contraseña incorrecta. Ingrese de nuevo:")
else:
  print ("Contraseña correcta.")
```

```python
def user_1(a,b):
  if a=="Juan" and b=="12345_":
    result= True
  else:
    result= False
  return result

def user_2(c,d):
  if c=="Pablo" and d=="xDcFvGbHn":
    result= True
  else:
    result= False
  return result

input_1=input("Ingrese su usuario: ")
input_2=input("Ingrese su contraseña: ")

while not user_1(input_1, input_2) and not user_2(input_1, input_2):
  input_1= input("Los datos ingresados son incorrectos. Ingrese su usuario nuevamente: ")
  input_2= input("Ingrese nuevamente su contraseña: ")
else:
  print("Ingreso exitoso.")
```

```python
#Versión mejorada usando solo una función. Inspirada en la solución.
def login(a,b):
  if (a=="Juan" and b=="12345_") or \
  (a=="Pablo" and b=="xDcFvGbHn"):
    result= True
  else:
    result= False
  return result

input_1=input("Ingrese su usuario: ")
input_2=input("Ingrese su contraseña: ")

while not login(input_1, input_2):
  input_1= input("Los datos ingresados son incorrectos. Ingrese su usuario nuevamente: ")
  input_2= input("Ingrese nuevamente su contraseña: ")
else:
  print("Ingreso exitoso.")
```

```python
def cdivisors(x):
  i=0
  for j in range(1,x+1):
    if x%j==0:
      i+=1
  return i

print("Calculadora de cantidad de divisores")
number= int(input("Ingrese un número: "))
print(cdivisors(number))
```
