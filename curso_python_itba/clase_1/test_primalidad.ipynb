```python
#Test usando lo ya escrito para divisores
def cdivisors(x):
  i=0
  for j in range(1,x+1):
    if x%j==0:
      i+=1
  return i


num=int(input("Ingrese un número: "))

if cdivisors(num)==2:
  print("¿Es primo?", "True")
else:
  print("¿Es primo?", "False")
```

```python
#Test usando criterio de la raíz (Criba de Eratóstenes)
#Solución propuesta
def esPrimo(n):
  if n == 1:
    return False
  
  d=2
  while d <= n**0.5:
    if n%d == 0:
      return False
    d = d+1            #La recursión se añade acá para que pruebe con todos los números antes de pasar al else. Es equivalente a poner continue.               
  else:              
      return True
  
for x in range(1,21):
  print(x, "¿Es primo?", esPrimo(x))
```
