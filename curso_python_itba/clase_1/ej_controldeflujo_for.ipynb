```python
#Notar que toma 0 como valor inicial y da saltos de a 1 por defecto. El 5 lo toma como fin.
print("range(5):")
for x in range(5):
    print(x)

#Acá toma los valores como inicio/fin respectivamente y da saltos de 1.
print("range(10, 15):")
for x in range(10, 15):
    print(x)
```

```python
print("Sistema de control de impresión de etiquetas")
n_prod=int(input("Ingrese la cantidad de productos a etiquetar: "))

for i in range(n_prod):
  prod=(input("Ingrese el código de barras: "))
  rep=int(input("Ingrese la cantidad de veces a imprimir: "))
  
  for j in range(rep):
    print(prod)
```
