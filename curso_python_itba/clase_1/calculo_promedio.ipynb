```python
#Parte 1
def promedio(a,b,c):
  return((a+b+c)/3)

nota1=(int(input("Ingrese la primer nota: ")))
nota2=(int(input("Ingrese la segunda nota: ")))
nota3=(int(input("Ingrese la tercer nota: ")))

print("El promedio es:", promedio(nota1,nota2,nota3))
```

```python
#Parte 2
def promedioponderado(a,b,c):
  return(0.2*a+0.5*b+0.3*c)

nota1=(int(input("Ingrese la primer nota: ")))
nota2=(int(input("Ingrese la segunda nota: ")))
nota3=(int(input("Ingrese la tercer nota: ")))

print("El promedio es:", promedioponderado(nota1,nota2,nota3))
```

```python
#Parte 3
def promedio(a,b,c):
  return((a+b+c)/3)

def promedioponderado(a,b,c):
  return(0.2*a+0.5*b+0.3*c)

for i in range(3):
  nota1=(int(input("Ingrese la primer nota: ")))
  nota2=(int(input("Ingrese la segunda nota: ")))
  nota3=(int(input("Ingrese la tercer nota: ")))

  if promedio(nota1,nota2,nota3)>=4:
    print("Aprobado en base al promedio común")
  else:
    print("Desaprobado en base al promedio común")
  
  if promedioponderado(nota1,nota2,nota3)>=4:
    print("Aprobado en base al promedio ponderado")
  else:
    print("Desaprobado en base al promedio ponderado")
```
