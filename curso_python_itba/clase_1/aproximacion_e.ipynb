```python
#Primero creo una función factorial
def factorial(n):
  if n==0:
    n+=1
  else:
    for i in range (1, n):
      n=n*i
  return n

nro=int(input("Ingrese un número no negativo: "))
factorial(nro)
```

```python
#Función factorial definida usando recursión (Anexo A)
def f(n):
  if n <= 2:
    return n
  else:
    return n * f( n-1 )       #la función se llama a sí misma.

x=int(input("Ingrese un número: "))
f(x)
```

```python
#Otra forma con recursión
def factorial(n):
  return 1 if n<=1 else n*factorial(n-1)

factorial(5)
```

```python
#Definición de factorial propuesta en el solucionario
def factorial(n):
  resultado = 1
  x = 2
  while x <= n:
    resultado *= x
    x += 1           #else implícito que hace resultado=1 para n=0 y n=1.
  return resultado

nro=int(input("Ingrese nro: "))
factorial(nro)
```
